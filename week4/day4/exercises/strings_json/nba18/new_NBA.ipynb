```python
import requests
import json
import pandas as pd
```

```python
r = requests.get(url='http://data.nba.net/prod/v2/2018/teams.json')
json_readed = r.json()
```

print(json_readed)

```python
with open ( "data_indented.json", "w+") as outfileNBA:
    json.dump(json_readed, outfileNBA, indent = 4)
```

```python
df_jsonNBA = pd.read_json("data_indented.json")
df_jsonNBA
```

```python
#Aquí estoy creando una variable que contiene lo que hay dentro de la columna "league" del dataframe (es decir, la lista "standard" que contiene el diccionario al que queremos acceder)
new_json = df_jsonNBA.league
```

```python
print(new_json)
```

```python
# Aquí estoy creando una variable qu contiene lo que hay dentro de "standard" (el diccionario que buscamos atrapar para concatenar después)
dentro_league = new_json ["standard"]

# Creamos un archivo .json en la carpeta para almacenarlo
with open ('dentro_league.json', 'w+') as outfile:
    json.dump(dentro_league, outfile, indent=4)

```

```python
# Creamos el dataframe del contenido de la columna league ya limpio
df_dentro_league = pd.read_json("dentro_league.json") 
```

```python
print(df_dentro_league)
```

```python
# Concatenamos los dos Dataframes con axis = 0
final_df = pd.concat([df_dentro_league, df_jsonNBA] , axis = 1, sort= False)
final_df
```
